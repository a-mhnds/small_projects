```python
import pandas as pd


df = pd.read_csv('avocado.csv', index_col=0)
df.info()
```

```python
df.Date = pd.to_datetime(df.Date)
df.index = df.Date
df.drop('Date', axis=1, inplace=True)
```

```python
df
```

```python
df.sort_index(inplace=True)
```

```python
df.AveragePrice.plot(figsize=(10,5))
```

```python
months = df.index.month
df['Month'] = months
df
```

```python
df.drop('year', axis=1, inplace=True)
```

```python
df.type = df.type.map({'organic': 0, 'conventional': 1})
```

```python
df.region.nunique()
```

One-hot encode all objects in the dataframe using the get_dummies in Pandas 

```python
df = pd.get_dummies(df)
```

```python
from sklearn.preprocessing import StandardScaler


y = df['AveragePrice']
X = df.drop('AveragePrice', axis=1)


X_train = X.iloc[ : int(0.8*len(X)), :]
X_test = X.iloc[int(0.8*len(X)) : , :]
X_val = X_test.iloc[: int(0.5*len(X_test)), :]
X_test = X_test.iloc[int(0.5*len(X_test)) : , :]


y_train = y[ : int(0.8*len(y))]
y_test = y[int(0.8*len(y)) : ]
y_val = y_test[: int(0.5*len(y_test))]
y_test = y_test[int(0.5*len(y_test)) : ]

scaler_train = StandardScaler()
scaler_val = StandardScaler()
X_train = scaler_train.fit_transform(X_train)
X_val = scaler_val.fit_transform(X_val)
```

```python
from sklearn.neighbors import KNeighborsRegressor

score_arr = []
for k in range(1,50):
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_train, y_train)

    score_arr.append(model.score(scaler_train.transform(X_test), y_test))



```

```python
import matplotlib.pyplot as plt

plt.scatter(range(len(score_arr)), score_arr)
```
